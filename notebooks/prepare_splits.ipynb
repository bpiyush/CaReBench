```python
import os
import sys

import pandas as pd
import shared.utils as su
```

```python
data_dir = "/scratch/shared/beegfs/piyush/datasets/SimCSE-NLI"

df_ego = pd.read_csv(f"{data_dir}/ego4d-422K-cleaned.csv")
df_nli = pd.read_csv(f"{data_dir}/nli-275k.csv")

def filter_csv(df): return df[~((df.sent0 == df.sent1) | (df.sent1 == df.hard_neg) | (df.sent0 == df.hard_neg))].reset_index(drop=True)

df_ego = filter_csv(df_ego)
df_nli = filter_csv(df_nli)
df_ego['source'] = 'ego4d'
df_nli['source'] = 'nli'

df_ego.shape, df_nli.shape
```

```python
fracs = [
    (0.0, 1.0),
    (0.05, 0.95),
    (0.1, 0.9),
    (0.2, 0.8),
    (0.3, 0.7),
    (0.4, 0.6),
    (0.5, 0.5),
    (0.6, 0.4),
    (0.7, 0.3),
    (0.8, 0.2),
    (0.9, 0.1),
    (1.0, 0.0),
]

n = 30000
for fx, fy in fracs:
    nx = int(fx * n)
    ny = int(fy * n)
    z = pd.concat([df_ego.sample(n=nx), df_nli.sample(n=ny)]).sample(frac=1.).reset_index(drop=True)
    
    save_path = f"{data_dir}/fractions/nli-{int(ny/1000)}k-ego4d-{int(nx/1000)}k.csv"
    os.makedirs(os.path.dirname(save_path), exist_ok=True)
    z.to_csv(save_path, index=False)
    print(save_path)
```
