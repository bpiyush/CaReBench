```python
import os
import torch
import numpy as np
```

```python
X = torch.randn((100, 512))


def compute_eff_dim(X, gamma=0.99): # X: N x D

    # Compute covariance matrix: D x D
    C = torch.cov(X.T)
    
    # Run SVD on C to get the singular values
    U, S, V = torch.linalg.svd(C)
    A = np.cumsum(S) / S.sum()
    
    d_eff = np.argmax(A > gamma).item()

    return d_eff


compute_eff_dim(X, 0.99)
```
